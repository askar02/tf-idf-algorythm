### 2. Each line is the one sentence. Please make them all in lower case form using string function lower(). EXAMPLE: in comparison to dogs, cats have not undergone major changes during the domestication process.

```python
a = open('words.txt', 'r')
sentence = []
for row in a:
    row = row.lower()
    sentence.append(row)
sentence
```

### 3. Tokenization. Means that splitting the sentences to the words. For that purpose you can use regular expressions, that can split the words by space or any other symbols that aren’t letters. re.split('[^a-z]', t). Do not forgot to remove empty words. EXAMPLE: ['in', 'comparison', 'to','dogs', '', 'cats', 'have', 'not', 'undergone', 'major', 'changes', 'during', 'the', 'domestication','process']

```python
words = []
for word in sentence:
    word = re.split('[^a-z]', word)
    words.append(word)
words
```

### 4. Make a list of all the words that appear in the sentences. Note: all the words are unique. And give the index to the each sentence index from 0 to #of_the_unique_words. You can use dict. Example: {0: ‘mac', 1: 'permanently',2: 'osx',3: 'download',4: 'between',5: ‘based', 6: ‘which', …………., 252: ‘safer', 253: ‘will’}. Hint: we have 254 unique words.

```python
unique_words = []
for i in words:
    for word in i:
        unique_words.append(word)
unique = list(set(unique_words))
unique.remove('')
len(unique)
```

```python
dict = {i : unique[i] for i in range(0, len(unique))}
dict
```

### 5. And create Matrix with N x D dimensions. N is the number of the sentences and D is the number of the unique words (22 x 254). Fill it in: the element with index (i, j) in this matrix must be equal to the number of occurrences of the j-th word in the i-th sentence. (bag of words) Example of bag of words:

```python
def num_of_words(s,w):
    s = re.split('[^a-z]', s)
    n = 0
    for i in s:
        if i == w:
            n = n + 1
    return n
```

```python
len(sentence)
```

```python
len(unique)
```

```python
i = len(unique)
j = len(sentence)
matrix = [[0 for x in range(i)] for y in range(j)]
```

```python
i = 0
j = 0
for s in sentence:
    for w in unique:
        matrix[i][j] = num_of_words(s,w)
        j = j+1
    j = 0
    i = i+1
```

```python
table = pd.DataFrame(data = matrix, columns = unique, index = sentence)
table
```

### 6. Find the cosine distance from first sentence to the all other sentences. Which two sentences is close to the first sentence. You can use scipy.spatial.distance.cosine

```python
from sklearn.metrics.pairwise import cosine_distances, cosine_similarity
```

```python
dist = []
for i in range(1,len(table)):
    a = cosine_distances(table.iloc[0:1], table.iloc[i:i+1])
    dist.append(a)
dist

print(sorted(dist)[0:2])

a = dist.index(sorted(dist)[0])
b = dist.index(sorted(dist)[1])
print(sentence[0])
print('First sentence:',sentence[a])
print('Second sentence:',sentence[b])
```
